```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.zeros((65,32256))
```

```python
for i in range(15,36):
    if(i<10):
        path = "/Users/mundogn/Documents/Estadistica-III-Proy1/CroppedYale_3/yaleB0"+str(i)+"/1.pgm"
    else:
        path = "/Users/mundogn/Documents/Estadistica-III-Proy1/CroppedYale_3/yaleB"+str(i)+"/1.pgm"
    im = Image.open(path)
    aux = np.asarray(im.getdata())
    aux = np.transpose(aux)
    X[i-15] = aux
```

```python
fig,axes = plt.subplots( 4,5, figsize=[12,9] )
for i,iax in enumerate( axes.flatten() ):
    im.putdata(X[i])
    iax.imshow(im)
fig.show()
```
